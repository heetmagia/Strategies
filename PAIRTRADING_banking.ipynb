```python
import pandas as pd

def high_corr_pairs(corr_df, threshold= 0.50):
    pairs = []
    
    for i in range(len(corr_df.columns)):
        for j in range(i + 1, len(corr_df.columns)):
            stock1 = corr_df.columns[i]
            stock2 = corr_df.columns[j]
            corr_value = corr_df.iloc[i, j]
            
            if abs(corr_value) >= threshold:
                pairs.append((stock1, stock2, corr_value))
    
    return pd.DataFrame(
        pairs, columns=["Stock_1", "Stock_2", "Correlation"]
    )

strong_pairs = high_corr_pairs(correlation_matrix, threshold=0.80)

strong_pairs.sort_values(
    by="Correlation",
    ascending=False,
    inplace=True
)

strong_pairs
```

```python
def get_high_corr_pairs(returns_df, list1, list2, threshold=0.50):
    """
    returns_df: pandas DataFrame of daily returns
    list1: first list of tickers
    list2: second list of tickers
    threshold: minimum absolute correlation to consider
    """
    pairs = []
    for i in list1:
        for j in list2:
            if i == j:
                continue  # skip same stock
            corr_val = returns_df[i].corr(returns_df[j])
            if abs(corr_val) >= threshold:
                pairs.append((i, j, corr_val))
    return pd.DataFrame(pairs, columns=["Stock_1", "Stock_2", "Correlation"])
```

```python
# Private ↔ Private (skip duplicates)
private_corr = get_high_corr_pairs(returns, PRIVATE_BANKS, PRIVATE_BANKS)
private_corr = private_corr[private_corr["Stock_1"] < private_corr["Stock_2"]]

# PSU ↔ PSU
psu_corr = get_high_corr_pairs(returns, PSU_BANKS, PSU_BANKS)
psu_corr = psu_corr[psu_corr["Stock_1"] < psu_corr["Stock_2"]]

# Private ↔ PSU
private_psu_corr = get_high_corr_pairs(returns, PRIVATE_BANKS, PSU_BANKS)
```
